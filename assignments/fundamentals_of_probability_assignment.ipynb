# Q1) The Probability of Shooting at a Target

A point is chosen at random from a disk of radius $10$. Let $A$ be the event that the point lies within $1$ unit of the boundary.

1. Model the experiment assuming the point is uniformly distributed with respect to area, and compute $P(A)$.
2. Model the experiment assuming the distance to the point from the center is chosen uniformly from $[0,10]$, and the direction is chosen independently and uniformly from $[0,2\pi)$ and compute $P(A)$.
3. Explain why the two answers differ.


Q1) The Probability of Shooting at a TargetLet the disk be defined as:$$\Omega = \{(x, y) \in \mathbb{R}^2 : x^2 + y^2 \leq 10^2\}$$Let $A$ be the event that the chosen point lies within $1$ unit of the boundary. Since the boundary is the circle of radius $10$, this means the point has a distance at least $9$ from the center and at most $10$. Thus:$$A = \{(x, y) \in \Omega : 9 \leq \sqrt{x^2 + y^2} \leq 10\}$$So $A$ is the outer annulus of the disk. We now model the phrase "chosen at random" in two different ways.1. Uniform distribution with respect to areaIn this model, equal areas are assigned equal probabilities. Thus, the probability of an event is its area divided by the total area of the disk.The total area of the disk is:$$\pi(10)^2 = 100\pi$$The event $A$ is the annulus between radii $9$ and $10$, so its area is:$$\pi(10)^2 - \pi(9)^2 = 100\pi - 81\pi = 19\pi$$Therefore,$$P(A) = \frac{\text{area of } A}{\text{area of disk}} = \frac{19\pi}{100\pi} = \frac{19}{100}$$Hence, under the area-uniform model,$$P(A) = 0.19$$2. Uniform distribution with respect to radiusIn this model, the distance from the center $r$ is chosen uniformly from the interval $[0, 10]$. The event $A$ occurs if $r \in [9, 10]$.Since $r \sim \text{Uniform}(0, 10)$, the probability is the length of the successful interval divided by the total length of the sample space:$$P(A) = \frac{\text{length of } [9, 10]}{\text{length of } [0, 10]} = \frac{10 - 9}{10 - 0} = \frac{1}{10}$$Hence, under the radius-uniform model,$$P(A) = 0.10$$3. Why the two answers differThe two answers differ because the probability density is distributed differently across the disk in each model:Area-Uniform: Recognizes that as you move further from the center, there is more physical area (more points) available. This model places higher weight on the outer sections of the circle.Radius-Uniform: Assumes every radial distance is equally likely. This effectively "crowds" the center of the disk because a small change in $r$ near the center covers much less area than the same change in $r$ near the boundary, yet both are given the same probability.

# Q2) Bertrand's Paradox

A chord of a circle is chosen “at random.” What is the probability that its length exceeds the side length of the inscribed equilateral triangle?

Show that different mathematically natural models of “random chord” lead to different answers.

Q2) Bertrand's ParadoxConsider a circle with radius $R$. An inscribed equilateral triangle has a side length of $L = R\sqrt{3}$. We want to find the probability $P(A)$ that a chord chosen "at random" has a length $X > R\sqrt{3}$.Bertrand's Paradox demonstrates that the answer depends on how the "random" selection is modeled.Model 1: Random EndpointsIn this model, we choose two points at random on the circumference of the circle and connect them to form a chord.Fix one endpoint of the chord at one vertex of the inscribed equilateral triangle.The chord will be longer than the side of the triangle if the second endpoint lies on the arc between the other two vertices.The circumference is divided into three equal arcs by the vertices of the triangle, each of length $\frac{2\pi R}{3}$.The "favorable" arc length is $\frac{1}{3}$ of the total circumference.Therefore,$$P(A) = \frac{1}{3}$$Model 2: Random Radial DistanceIn this model, we choose a radius of the circle and then choose a point at random along that radius to serve as the midpoint of the chord.The side of an inscribed equilateral triangle is at a distance of $r = \frac{R}{2}$ from the center.A chord is longer than the side of the triangle if its midpoint is closer to the center than the midpoint of the triangle's side.The total length of the radius is $R$.The "favorable" distance for the midpoint is the interval $[0, R/2]$.Therefore,$$P(A) = \frac{R/2}{R} = \frac{1}{2}$$Model 3: Random MidpointIn this model, we choose a point at random anywhere within the disk to be the midpoint of the chord.A point uniquely determines a chord (except for the center).The chord is longer than the side of the triangle if the chosen midpoint falls within a smaller concentric circle that is tangent to the sides of the triangle.The radius of this inner circle is $r = \frac{R}{2}$.The area of the disk is $\pi R^2$.The "favorable" area is $\pi (R/2)^2 = \frac{\pi R^2}{4}$.Therefore,$$P(A) = \frac{\text{Area of inner circle}}{\text{Area of disk}} = \frac{\pi R^2 / 4}{\pi R^2} = \frac{1}{4}$$

#Q8) Probability of Being a Girl?

A family has two children. Assume each child is equally likely to be a boy or a girl, independently of the other.

What is the probability that both children are girls, given that **at least one** of them is a girl?

Q8) Probability of Being a Girl?To find the probability that both children are girls given that at least one of them is a girl, we must first define the sample space and the relevant events.1. Define the Sample SpaceA family has two children. Let B represent a boy and G represent a girl. The possible outcomes for the two children (ordered by birth) are:$$\Omega = \{BB, BG, GB, GG\}$$Since each child is equally likely to be a boy or a girl independently, each outcome in $\Omega$ has a probability of:$$P(\text{each outcome}) = \frac{1}{2} \times \frac{1}{2} = \frac{1}{4}$$2. Identify the EventsLet Event $A$ be the condition that both children are girls:$$A = \{GG\}$$Let Event $L$ be the condition that at least one child is a girl:$$L = \{BG, GB, GG\}$$3. Calculate the Conditional ProbabilityWe want to find the probability of $A$ given $L$, denoted as $P(A|L)$. According to the formula for conditional probability:$$P(A|L) = \frac{P(A \cap L)}{P(L)}$$Step A: Find $P(A \cap L)$The intersection $A \cap L$ represents the outcomes where both are girls and at least one is a girl. This is simply the outcome $\{GG\}$.$$P(A \cap L) = \frac{1}{4}$$Step B: Find $P(L)$Event $L$ contains 3 out of the 4 possible outcomes.$$P(L) = \frac{3}{4}$$Step C: Final Calculation$$P(A|L) = \frac{1/4}{3/4} = \frac{1}{3}$$

A chemical plant monitors each production run using two different measurement systems.

For a given run, let

* $T$ be the deviation of the reactor temperature from target, measured in standardized units,
* $P$ be the deviation of the reactor pressure from target, also measured in standardized units.

Q10) Information Available in Continuous Random Variables
1. Probability Space $(\Omega, \mathfrak{F}, P)$Sample Space: $\Omega = \mathbb{R}^2$.$\sigma$-algebra: $\mathfrak{F} = \mathfrak{B}(\mathbb{R}^2)$ (the Borel $\sigma$-algebra on $\mathbb{R}^2$).Probability Measure $P$: The measure induced by the joint density function:$$f_{T,P}(t,p) = \frac{1}{4\pi} \exp\left( -\frac{t^2}{2} - \frac{p^2}{8} \right)$$
2. $\sigma$-algebra generated by $X$, $\sigma(X)$$X(t,p) = t$. This random variable only "sees" the first coordinate.$$\sigma(X) = \{ B \times \mathbb{R} : B \in \mathfrak{B}(\mathbb{R}) \}$$This represents all vertical strips in the 2D plane.
3. $\sigma$-algebra generated by $Y$, $\sigma(Y)$$Y(t,p) = (t,p)$ is the identity map. It "sees" both coordinates.$$\sigma(Y) = \mathfrak{B}(\mathbb{R}^2)$$
4. Show $\sigma(X) \subseteq \sigma(Y)$Since every set of the form $B \times \mathbb{R}$ is a Borel set in $\mathbb{R}^2$, $\sigma(X)$ is a sub-$\sigma$-algebra of $\sigma(Y)$. $Y$ contains more information because it specifies the exact point $(t,p)$, whereas $X$ only specifies the line $T=t$.  
5. Marginal Density of $X$Since $T \sim N(0,1)$:$$f_X(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}$$
6. Marginal Density of $Y$$Y$ is the joint pair $(T,P)$. The density is the joint PDF provided:$$f_Y(t,p) = \frac{1}{\sqrt{2\pi}(1)} e^{-\frac{t^2}{2}} \cdot \frac{1}{\sqrt{2\pi}(2)} e^{-\frac{p^2}{2(4)}} = \frac{1}{4\pi} \exp\left( -\frac{t^2}{2} - \frac{p^2}{8} \right)$$
7. Probabilities$P(X \leq 0) = \Phi(0) = 0.5$$P(X > 1) = 1 - \Phi(1) \approx 1 - 0.8413 = 0.1587$$P(Y \in (-\infty, 0] \times \mathbb{R}) = P(T \leq 0) = 0.5$$P(Y \in [-1, 1] \times [-2, 2]) = P(-1 \leq T \leq 1) \times P(-2 \leq P \leq 2)$$= [2\Phi(1) - 1] \times [2\Phi(1) - 1] \approx 0.6827^2 \approx 0.466$   

8. Conditional Distribution of $X$ given $Y=(t,p)$Since $X$ is a function of $Y$ ($X=t$), the distribution is a Dirac delta (point mass):$$f_{X|Y}(x | t,p) = \delta(x - t)$$
9. Conditional Distribution of $Y$ given $X=t$Since $T$ and $P$ are independent, knowing $T=t$ tells us nothing about $P$:$$f_{Y|X}(t, p | t) = f_P(p) = \frac{1}{\sqrt{8\pi}} e^{-\frac{p^2}{8}}$$
10. Information and ConditioningConditioning on $X$ fixes the temperature but leaves the pressure $P$ unknown. Conditioning on $Y$ fixes both, removing all uncertainty.   

11. $X = \pi_1 \circ Y$Since $X(t,p) = t$ and $Y(t,p) = (t,p)$, then $X = \pi_1(Y)$. Since the projection map $\pi_1$ is continuous, it is Borel-measurable, implying $X$ is a measurable function of $Y$.

13. Differential Entropy $h(X)$For $X \sim N(0, \sigma^2)$, $h(X) = \frac{1}{2}\ln(2\pi e \sigma^2)$.With $\sigma^2=1$: $h(X) = \frac{1}{2}\ln(2\pi e) \approx 1.419$ nats.
14. Differential Entropy $h(Y)$For independent variables, $h(T,P) = h(T) + h(P)$.$h(P) = \frac{1}{2}\ln(2\pi e \cdot 4) = \frac{1}{2}\ln(8\pi e)$.$h(Y) = \frac{1}{2}\ln(2\pi e) + \frac{1}{2}\ln(8\pi e) = \ln(4\pi e) \approx 3.528$ nats.

15. Conditional Differential Entropy $h(Y|X)$Because of independence, $h(Y|X) = h(P) = \frac{1}{2}\ln(8\pi e)$. Physically, this is the uncertainty remaining in the system (the pressure) once the temperature is known.
16. $h(X|Y)$In the continuous case, if $X$ is a function of $Y$, $h(X|Y) \to -\infty$. This differs from the discrete case (where $H(X|Y)=0$) because the precision of a continuous variable is infinite, making the "certainty" of a known value infinitely informative.
