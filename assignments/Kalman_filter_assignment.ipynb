# Q. Analytical Derivation

Consider the filter model
\begin{align*}
x^{-}_k &= A_{k-1}\,x^{+}_{k-1} + G_{k-1}\,w_{k-1}, \\
y^{-}_k &= H_k\,x^{-}_k + z_k,
\end{align*}
where
$$x^{+}_{k-1} \sim \mathscr{N}(m_{k-1}, P_{k-1}),$$

Answer the followings:

1. Show that
$$x_k^- \sim \mathscr{N}(m_k^-, P_k^-),$$
where
$$m_k^- = A_{k-1}m_{k-1},$$
and
$$
P_k^- = A_{k-1}P_{k-1}A_{k-1}^T + G_{k-1}\Sigma_p G_{k-1}^T.
$$

2. Show that
$$y_k^- \sim \mathscr{N}(H_k m_k^-, H_k P_k^- H_k^T + \Sigma_m).$$

3. Show that
$$\begin{bmatrix} x_k^- \\ y^{-}_k \end{bmatrix} \sim \mathscr{N}\left(
\begin{bmatrix} m_k^- \\ H_k m_k^- \end{bmatrix},
\begin{bmatrix} P_k^- & P_k^- H_k^T \\ H_k P_k^- & H_k P_k^- H_k^T + \Sigma_m \end{bmatrix}
\right).$$

4. Show that
$$x^{+}_k\triangleq (x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}) \sim \mathscr{N}(m_k, P_k),$$
where the updated mean $m_k$ and updated covariance $P_k$ are given by:
\begin{align*}
K_k &\triangleq P_k^- H_k^T (H_k P_k^- H_k^T + \Sigma_m)^{-1}, \\
m_k &= m_k^- + K_k (y^{\mathrm{obs}}_{k} - H_k m_k^-), \\
P_k &= (I - K_k H_k) P_k^-.
\end{align*}

5. Find
$\mathbb{E}[x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k}]$  and
$\text{Var}(x^{-}_k \mid y^{-}_k = y^{\mathrm{obs}}_{k})$.

1. State Prediction Derivation
Given the system model $x_{k}^{-}=A_{k-1}x_{k-1}^{+}+G_{k-1}w_{k-1}$, we know that linear transformations of Gaussian random variables result in a Gaussian distribution. Since the prior state $x_{k-1}^{+}$ and the process noise $w_{k-1}$ are independent zero-mean Gaussians, the predicted state $x_{k}^{-}$ is also Gaussian.  To find its parameters:$$E[x_{k}^{-}] = E[A_{k-1}x_{k-1}^{+} + G_{k-1}w_{k-1}] = A_{k-1}E[x_{k-1}^{+}] + G_{k-1}E[w_{k-1}]$$Since the expected value of the process noise is zero, this simplifies to the predicted mean:
  $$m_{k}^{-} = A_{k-1}m_{k-1}$$For the covariance, assuming state and noise are uncorrelated:$$Var(x_{k}^{-}) = Var(A_{k-1}x_{k-1}^{+}) + Var(G_{k-1}w_{k-1})$$$$P_{k}^{-} = A_{k-1}P_{k-1}A_{k-1}^{T} + G_{k-1}\Sigma_{p}G_{k-1}^{T}$$This confirms that $x_{k}^{-}\sim\mathcal{N}(m_{k}^{-},P_{k}^{-})$.
 2. Given Assumptions:The measurement model is defined as $y_{k}^{-}=H_{k}x_{k}^{-}+z_{k}$.The predicted state is a Gaussian distribution: $x_{k}^{-}\sim\mathcal{N}(m_{k}^{-},P_{k}^{-})$.The measurement noise is a zero-mean Gaussian distribution: $z_{k}\sim\mathcal{N}(0,\Sigma_{m})$.The state $x_{k}^{-}$ and the measurement noise $z_{k}$ are statistically independent.Because $y_{k}^{-}$ is a linear combination of two Gaussian random variables ($x_{k}^{-}$ and $z_{k}$), the resulting distribution of $y_{k}^{-}$ must also be strictly Gaussian. To fully define this new Gaussian distribution, we just need to find its expected value (mean) and variance (covariance).Step 1: Derive the Expected Value (Mean)Take the expectation of both sides of the measurement equation:$$\mathbb{E}[y_{k}^{-}]=\mathbb{E}[H_{k}x_{k}^{-}+z_{k}]$$By the linearity property of expectations, we can separate the terms:$$\mathbb{E}[y_{k}^{-}]=H_{k}\mathbb{E}[x_{k}^{-}]+\mathbb{E}[z_{k}]$$Substitute the known expected values ($\mathbb{E}[x_{k}^{-}]=m_{k}^{-}$ and $\mathbb{E}[z_{k}]=0$):$$\mathbb{E}[y_{k}^{-}]=H_{k}m_{k}^{-}+0$$$$\mathbb{E}[y_{k}^{-}]=H_{k}m_{k}^{-}$$Step 2: Derive the CovarianceTake the variance of both sides of the measurement equation:$$Var(y_{k}^{-})=Var(H_{k}x_{k}^{-}+z_{k})$$Because the state estimate and the measurement noise are independent, the variance of their sum is equal to the sum of their individual variances:$$Var(y_{k}^{-})=Var(H_{k}x_{k}^{-})+Var(z_{k})$$Applying the standard covariance property for linear transformations ($Var(AX)=A Var(X)A^{T}$), we pull the $H_{k}$ matrix out:$$Var(y_{k}^{-})=H_{k}Var(x_{k}^{-})H_{k}^{T}+Var(z_{k})$$Substitute the known covariances ($Var(x_{k}^{-})=P_{k}^{-}$ and $Var(z_{k})=\Sigma_{m}$):$$Var(y_{k}^{-})=H_{k}P_{k}^{-}H_{k}^{T}+\Sigma_{m}$$Step 3: ConclusionSince $y_{k}^{-}$ is a Gaussian random variable defined by the mean and covariance derived above, we have successfully shown that:$$y_{k}^{-}\sim\mathcal{N}(H_{k}m_{k}^{-},H_{k}P_{k}^{-}H_{k}^{T}+\Sigma_{m})$$
3. Measurement Prediction Derivation
Given the measurement model $y_{k}^{-}=H_{k}x_{k}^{-}+z_{k}$ , we find the expected value and variance similarly:
  $$E[y_{k}^{-}] = E[H_{k}x_{k}^{-} + z_{k}] = H_{k}E[x_{k}^{-}] + 0 = H_{k}m_{k}^{-}$$$$Var(y_{k}^{-}) = Var(H_{k}x_{k}^{-}) + Var(z_{k}) = H_{k}P_{k}^{-}H_{k}^{T} + \Sigma_{m}$$To find the joint distribution, we must calculate the cross-covariance between the state and the measurement:
  $$\Sigma_{xy} = E[(x_{k}^{-} - m_{k}^{-})(y_{k}^{-} - H_{k}m_{k}^{-})^{T}]$$$$\Sigma_{xy} = E[(x_{k}^{-} - m_{k}^{-})(H_{k}(x_{k}^{-} - m_{k}^{-}) + z_{k})^{T}]$$$$\Sigma_{xy} = E[(x_{k}^{-} - m_{k}^{-})(x_{k}^{-} - m_{k}^{-})^{T}]H_{k}^{T} + E[(x_{k}^{-} - m_{k}^{-})z_{k}^{T}]$$
Because state error and measurement noise are independent, the second term is zero, leaving $\Sigma_{xy} = P_{k}^{-}H_{k}^{T}$. This forms the upper right block of the joint covariance matrix, yielding the full joint distribution.
 4. Posterior State Update (Conditioning)
The updated state $x_{k}^{+}$ is the conditional distribution $(x_{k}^{-}|y_{k}^{-}=y_{k}^{obs})$. Using the standard theorem for jointly Gaussian variables, the conditional mean and covariance are given by:
  $$\mu_{x|y} = \mu_{x} + \Sigma_{xy}\Sigma_{yy}^{-1}(y - \mu_{y})$$$$\Sigma_{x|y} = \Sigma_{xx} - \Sigma_{xy}\Sigma_{yy}^{-1}\Sigma_{yx}$$Substituting the blocks from the joint distribution derived in Step 3:  $\mu_{x} = m_{k}^{-}$$\mu_{y} = H_{k}m_{k}^{-}$$\Sigma_{xy} = P_{k}^{-}H_{k}^{T}$$\Sigma_{yy} = H_{k}P_{k}^{-}H_{k}^{T} + \Sigma_{m}$Defining the Kalman Gain $K_{k} = \Sigma_{xy}\Sigma_{yy}^{-1} = P_{k}^{-}H_{k}^{T}(H_{k}P_{k}^{-}H_{k}^{T} + \Sigma_{m})^{-1}$ , the updated mean is:
  $$m_{k} = m_{k}^{-} + K_{k}(y_{k}^{obs} - H_{k}m_{k}^{-})$$And the updated covariance is:
  $$P_{k} = P_{k}^{-} - K_{k}H_{k}P_{k}^{-} = (I - K_{k}H_{k})P_{k}^{-}$$5. Expected Value and Variance
The terms $\mathbb{E}[x_{k}^{-}|y_{k}^{-}=y_{k}^{obs}]$ and $Var(x_{k}^{-}|y_{k}^{-}=y_{k}^{obs})$  represent the mean and variance of the state conditioned on the new measurement. These are precisely the updated mean $m_{k}$ and covariance $P_{k}$ derived in Step 4.

# Q. 1-D Example

Consider the scalar linear-Gaussian filter model
\begin{aligned}
x^-_k &= a\,x^+_{k-1} + w_{k-1},\qquad w_{k-1}\sim\mathscr N(0,q),\\
y^-_k &= h\,x^-_k + z_k,\qquad\;\;\;\; z_k\sim\mathscr N(0,r),
\end{aligned}
with prior $x_0\sim \mathscr{N}(m_0,P_0)$. Define $Y_{k}=\{y^{\mathrm{obs}}_1,\dots,y^{\mathrm{obs}}_{k}\}$.

Answer the following:

1. Show that
\begin{aligned}
m_k^- &= a\,m_{k-1},\\
P_k^- &= a^2 P_{k-1} + q.
\end{aligned}

2. Show that
\begin{aligned}
m_k &= m_k^- + K_k\,v_k
= m_k^- + \frac{P_k^- h}{S_k}\bigl(y^{\mathrm{obs}}_k - h\,m_k^-\bigr),\\
P_k &= (1 - K_k h)\,P_k^-
= \Bigl(1 - \frac{P_k^- h^2}{S_k}\Bigr) P_k^- .
\end{aligned}

3. Show that the predictive measurement distribution (before seeing $y_k$):
\begin{align*}
p(y^-_k | Y_{k-1})=\mathscr N\bigl(h\,m_k^-,\; h^2 P_k^- + r\bigr).
\end{align*}

4. Show that the posterior-predictive measurement distribution (after filtering on $y_k$):
\begin{align*}
p(y^-_k\mid Y_k)=\mathscr N\bigl(h\,m_k,\; h^2 P_k + r\bigr).
\end{align*}

5. For suitable numerical values for $a,q,h,r$ animate the above to prior distribution and the posterior distribution.

For a scalar (1-D) linear-Gaussian filter, all matrices from the general Kalman Filter become simple scalar values. Based on the provided model:State equation: $x_{k}^{-} = a x_{k-1}^{+} + w_{k-1}$ where $w_{k-1}\sim\mathcal{N}(0,q)$.  Measurement equation: $y_{k}^{-} = hx_{k}^{-} + z_{k}$ where $z_{k}\sim\mathcal{N}(0,r)$.  
1. Show State Prediction EquationsObjective: Show that $m_{k}^{-} = a m_{k-1}$ and $P_{k}^{-} = a^{2}P_{k-1} + q$.  Proof:Take the expected value (mean) of the state equation:$$\mathbb{E}[x_{k}^{-}] = \mathbb{E}[a x_{k-1}^{+} + w_{k-1}]$$Since the expected value of the noise $\mathbb{E}[w_{k-1}] = 0$:$$m_{k}^{-} = a \mathbb{E}[x_{k-1}^{+}] + 0 = a m_{k-1}$$Next, take the variance of the state equation. Because the previous state and the process noise are independent, their variances add:$$Var(x_{k}^{-}) = Var(a x_{k-1}^{+}) + Var(w_{k-1})$$Using the scalar variance property $Var(aX) = a^2Var(X)$:$$P_{k}^{-} = a^{2} P_{k-1} + q$$
2. Show State Update EquationsObjective: Show the 1D update steps for $m_k$ and $P_k$.  Proof:Substitute the 1D scalars into the general Kalman Filter update equations. The measurement matrix $H_k$ becomes $h$, and the measurement noise covariance $\Sigma_m$ becomes $r$.The innovation covariance (or scalar $S_k$) is:$$S_k = h P_k^- h^T + r = h^2 P_k^- + r$$The Kalman Gain ($K_k$) becomes:$$K_k = P_k^- h^T S_k^{-1} = \frac{P_k^- h}{S_k}$$The updated mean is:$$m_k = m_k^- + K_k (y_k^{obs} - h m_k^-)$$The updated covariance is:$$P_k = (1 - K_k h) P_k^- = \left(1 - \frac{P_k^- h^2}{S_k}\right) P_k^-$$
3. Show Predictive Measurement DistributionObjective: Show that $p(y_{k}^{-}|Y_{k-1}) = \mathcal{N}(h m_{k}^{-}, h^{2}P_{k}^{-} + r)$.  Proof:
This evaluates the measurement prediction before seeing the actual measurement $y_k$. Using the measurement equation $y_{k}^{-} = hx_{k}^{-} + z_{k}$:
  $$\mathbb{E}[y_{k}^{-}] = h\mathbb{E}[x_{k}^{-}] + \mathbb{E}[z_{k}] = h m_{k}^{-} + 0 = h m_{k}^{-}$$$$Var(y_{k}^{-}) = Var(hx_{k}^{-}) + Var(z_{k}) = h^{2}P_{k}^{-} + r$$Because it is a linear combination of Gaussians, the distribution is: $\mathcal{N}(h m_{k}^{-}, h^{2}P_{k}^{-} + r)$.

4. Show Posterior-Predictive Measurement DistributionObjective: Show that $p(y_{k}^{-}|Y_{k}) = \mathcal{N}(h m_{k}, h^{2}P_{k} + r)$.  Proof:
This asks for the measurement prediction after filtering on $y_k$. Once we have processed $y_k$, our state estimate updates from the prior ($m_k^-, P_k^-$) to the posterior ($m_k, P_k$).  If we were to predict a new measurement based on this newly updated state, we substitute the updated state mean and variance into the exact same expectation and variance operations used in Step 3:$$\mathbb{E}[y_{k}^{-}|Y_{k}] = h m_{k}$$$$Var(y_{k}^{-}|Y_{k}) = h^{2}P_{k} + r$$Resulting in the distribution: $\mathcal{N}(h m_{k}, h^{2}P_{k} + r)$.
5. Python Animation for ColabObjective: For suitable numerical values for $a$, $q$, $h$, $r$, animate the prior distribution and the posterior distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# 1. Define suitable numerical parameters
a = 1.0   # State transition scalar
q = 0.5   # Process noise variance
h = 1.0   # Observation scalar
r = 1.2   # Measurement noise variance

# Initial prior state parameters
m_prev = 0.0
P_prev = 1.0
y_obs = 3.0  # Simulated observation

# 2. Calculate Prior (Prediction Step)
m_prior = a * m_prev
P_prior = (a**2) * P_prev + q

# 3. Calculate Posterior (Update Step)
S = (h**2) * P_prior + r
K = (P_prior * h) / S
m_post = m_prior + K * (y_obs - h * m_prior)
P_post = (1 - K * h) * P_prior

# 4. Set up the figure for animation
x = np.linspace(-5, 7, 500)
fig, ax = plt.subplots(figsize=(8, 5))
ax.set_xlim(-5, 7)
ax.set_ylim(0, 0.6)
line, = ax.plot([], [], lw=2)
plt.close(fig) # Prevent static plot from showing in Colab

# 5. Animation function
def animate(frame):
    # Smoothly interpolate between prior (0) and posterior (100)
    alpha = frame / 100.0
    current_m = (1 - alpha) * m_prior + alpha * m_post
    current_P = (1 - alpha) * P_prior + alpha * P_post

    y = stats.norm.pdf(x, current_m, np.sqrt(current_P))
    line.set_data(x, y)

    if frame == 0:
        ax.set_title("Prior Distribution (Before Observation)")
        line.set_color('blue')
        line.set_label('Prior')
    elif frame == 100:
        ax.set_title(f"Posterior Distribution (After observing y={y_obs})")
        line.set_color('green')
        line.set_label('Posterior')
    else:
        ax.set_title("Bayesian Update Transition...")
        line.set_color('purple')

    return line,

# Render animation in Colab
anim = FuncAnimation(fig, animate, frames=101, interval=30, blit=True)
HTML(anim.to_jshtml())
```

# Q. 2D-Position Estimation

## Part A

Consider a two-dimensional constant-velocity dynamical system. The hidden state at time step $k$ is

$$
x_k =
\begin{bmatrix}
p_x(k)\\
p_y(k)\\
v_x(k)\\
v_y(k)
\end{bmatrix},
$$

where $(p_x(k),p_y(k))$ denote the position components and $(v_x(k),v_y(k))$ denote the velocity components.

The measurement consists only of the two position components:

$$
y_k =
\begin{bmatrix}
p_x^{\mathrm{meas}}(k)\\
p_y^{\mathrm{meas}}(k)
\end{bmatrix}.
$$

The linear Gaussian state-space filter model is

$$
x^-_k = A x^+_{k-1} + G w_{k-1},
$$

$$
y^-_k = Hx^+_k + z_k,
$$

where

$$
w_{k-1} \sim \mathscr{N}(0,\Sigma_p),
\qquad
z_k \sim \mathscr{N}(0,\Sigma_m).
$$

The process noise sequence $w_k$, measurement noise sequence $z_k$, and the initial state are assumed mutually independent.

Show that:

$$
A =
\begin{bmatrix}
1 & 0 & \Delta t & 0\\
0 & 1 & 0 & \Delta t\\
0 & 0 & 1 & 0\\
0 & 0 & 0 & 1
\end{bmatrix},
$$



$$
H =
\begin{bmatrix}
1 & 0 & 0 & 0\\
0 & 1 & 0 & 0
\end{bmatrix},
$$

and

$$
G =
\begin{bmatrix}
\frac{1}{2}\Delta t^2 & 0\\
0 & \frac{1}{2}\Delta t^2\\
\Delta t & 0\\
0 & \Delta t
\end{bmatrix}.
$$

For a 2D constant velocity model, the state vector is $x_{k}=[p_{x}(k), p_{y}(k), v_{x}(k), v_{y}(k)]^{T}$. By applying basic Newtonian kinematics, the position and velocity at time $k$ given a time step $\Delta t$ and acceleration noise $w$ (where $w_{x}$ and $w_{y}$ affect both velocity and position) are:
  $$p_{x}(k) = p_{x}(k-1) + v_{x}(k-1)\Delta t + \frac{1}{2}w_{x}(k-1)\Delta t^{2}$$$$v_{x}(k) = v_{x}(k-1) + w_{x}(k-1)\Delta t$$(The same equations apply identically for the $y$ dimension).Writing these equations in the matrix form $x_{k}^{-} = Ax_{k-1}^{+} + Gw_{k-1}$  yields the exact matrices provided:
  $$A = \begin{bmatrix} 1 & 0 & \Delta t & 0 \\ 0 & 1 & 0 & \Delta t \\ 0 & 0 & 1 & 0 \\ 0 & 0 & 0 & 1 \end{bmatrix}$$$$G = \begin{bmatrix} \frac{1}{2}\Delta t^{2} & 0 \\ 0 & \frac{1}{2}\Delta t^{2} \\ \Delta t & 0 \\ 0 & \Delta t \end{bmatrix}$$Since our sensor only measures the two position components $(p_{x}, p_{y})$, the observation matrix $H$ simply extracts the first two elements of the state vector:
  $$H = \begin{bmatrix} 1 & 0 & 0 & 0 \\ 0 & 1 & 0 & 0 \end{bmatrix}$$

## Part B

Develop a python based method to filter a sequence of noisy GPS obtained position meaurements.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Setup Parameters
dt = 1.0 # Time step
num_steps = 50

# State transition matrix A
A = np.array([[1, 0, dt, 0],
              [0, 1, 0, dt],
              [0, 0, 1, 0],
              [0, 0, 0, 1]])

# Observation matrix H
H = np.array([[1, 0, 0, 0],
              [0, 1, 0, 0]])

# Noise matrices
sigma_p = np.array([[0.1, 0], [0, 0.1]]) # Process noise variance
sigma_m = np.array([[5.0, 0], [0, 5.0]]) # Measurement noise variance (GPS is noisy)

# Process noise mapping matrix G
G = np.array([[0.5 * dt**2, 0],
              [0, 0.5 * dt**2],
              [dt, 0],
              [0, dt]])

Q = G @ sigma_p @ G.T # Full process noise covariance
R = sigma_m           # Measurement noise covariance

# 2. Generate Synthetic "True" Data and Noisy GPS Measurements
true_states = []
measurements = []
current_true_state = np.array([0.0, 0.0, 2.0, 1.0]) # [px, py, vx, vy]

for _ in range(num_steps):
    # Evolve true state (constant velocity + small random acceleration)
    accel_noise = np.random.multivariate_normal([0, 0], sigma_p)
    current_true_state = A @ current_true_state + G @ accel_noise
    true_states.append(current_true_state)

    # Generate noisy measurement
    meas_noise = np.random.multivariate_normal([0, 0], R)
    measurement = H @ current_true_state + meas_noise
    measurements.append(measurement)

true_states = np.array(true_states)
measurements = np.array(measurements)

# 3. Kalman Filter Implementation
# Initial guesses
m_k = np.array([0.0, 0.0, 0.0, 0.0]) # Initial state estimate
P_k = np.eye(4) * 10                 # Initial high uncertainty

filtered_states = []

for z in measurements:
    # --- Prediction Step ---
    m_prior = A @ m_k
    P_prior = A @ P_k @ A.T + Q

    # --- Update Step ---
    # S = H * P_prior * H^T + R
    S = H @ P_prior @ H.T + R
    # K = P_prior * H^T * S^-1
    K = P_prior @ H.T @ np.linalg.inv(S)

    # m_k = m_prior + K * (z - H * m_prior)
    y_residual = z - (H @ m_prior)
    m_k = m_prior + K @ y_residual

    # P_k = (I - K * H) * P_prior
    P_k = (np.eye(4) - K @ H) @ P_prior

    filtered_states.append(m_k)

filtered_states = np.array(filtered_states)

# 4. Plotting Results
plt.figure(figsize=(10, 6))
plt.plot(true_states[:, 0], true_states[:, 1], label='True Trajectory', color='green', linewidth=2)
plt.scatter(measurements[:, 0], measurements[:, 1], label='Noisy GPS Measurements', color='red', marker='x', alpha=0.6)
plt.plot(filtered_states[:, 0], filtered_states[:, 1], label='Kalman Filter Estimate', color='blue', linewidth=2, linestyle='--')

plt.title('2D Position Estimation via Kalman Filter')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.legend()
plt.grid(True)
plt.show()
```
