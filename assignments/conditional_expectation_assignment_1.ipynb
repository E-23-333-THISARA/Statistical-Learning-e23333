#Q1) Prediction of fraudulent transactions

A bank monitors credit-card transactions for possible fraud.

For each transaction, let the sample space be

$$
\Omega=\{\omega_1,\omega_2,\omega_3,\omega_4\},
$$

where:

* $\omega_1$: low-risk customer, legitimate transaction,
* $\omega_2$: low-risk customer, fraudulent transaction,
* $\omega_3$: high-risk customer, legitimate transaction,
* $\omega_4$: high-risk customer, fraudulent transaction.

Assume the probabilities are

$$
P(\{\omega_1\})=0.50,\qquad
P(\{\omega_2\})=0.10,\qquad
P(\{\omega_3\})=0.20,\qquad
P(\{\omega_4\})=0.20.
$$

Let $X:\Omega\to\mathbb R$ denote the fraud indicator:

$$
X(\omega_1)=0,\qquad
X(\omega_2)=1,\qquad
X(\omega_3)=0,\qquad
X(\omega_4)=1.
$$

Thus $X=1$ means the transaction is fraudulent, and $X=0$ means it is legitimate.

Now suppose the bank does not observe the full outcome $\omega$, but only the customer's risk class. Define $Y:\Omega\to\mathbb R$ by

$$
Y(\omega_1)=0,\qquad
Y(\omega_2)=0,\qquad
Y(\omega_3)=1,\qquad
Y(\omega_4)=1.
$$

Thus:

* $Y=0$ means “low-risk customer,”
* $Y=1$ means “high-risk customer.”

Answer the following.

1. Write down the probability space $(\Omega,\mathfrak F,P)$, where $\mathfrak F=\mathfrak P(\Omega)$.


Sample Space ($\Omega$):$$\Omega = \{\omega_1, \omega_2, \omega_3, \omega_4\}$$$\sigma$-algebra ($\mathfrak{F}$): Given as the power set $\mathfrak{P}(\Omega)$, which contains all $2^4 = 16$ possible subsets of $\Omega$:$$\mathfrak{F} = \{\emptyset, \{\omega_1\}, \{\omega_2\}, \{\omega_3\}, \{\omega_4\}, \{\omega_1, \omega_2\}, \{\omega_1, \omega_3\}, \{\omega_1, \omega_4\}, \{\omega_2, \omega_3\}, \{\omega_2, \omega_4\}, \{\omega_3, \omega_4\}, \{\omega_1, \omega_2, \omega_3\}, \{\omega_1, \omega_2, \omega_4\}, \{\omega_1, \omega_3, \omega_4\}, \{\omega_2, \omega_3, \omega_4\}, \Omega\}$$Probability Measure ($P$): We are given $P(\{\omega_1\}) = 0.50$, $P(\{\omega_2\}) = 0.10$, and $P(\{\omega_3\}) = 0.20$. Since the total probability must equal 1:$$P(\{\omega_4\}) = 1 - (0.50 + 0.10 + 0.20) = 0.20$$


2. Compute the $\sigma$-algebra generated by $Y$, namely $\sigma(Y)$.


The random variable $Y$ groups the outcomes by risk level:$Y^{-1}(0) = \{\omega_1, \omega_2\} = A_0$ (Low risk)$Y^{-1}(1) = \{\omega_3, \omega_4\} = A_1$ (High risk)The $\sigma$-algebra $\sigma(Y)$ is the smallest collection of sets containing these pre-images, closed under complement and countable union:$$\sigma(Y) = \{\emptyset, \{\omega_1, \omega_2\}, \{\omega_3, \omega_4\}, \Omega\}$$

3. Explain why $E[X\mid Y]$ must be $\sigma(Y)$-measurable, and therefore must be constant on the sets $\{ \omega_1,\omega_2\}$ and $\{\omega_3,\omega_4\}$.

By definition, a conditional expectation $\mathbb{E}[X|Y]$ is a random variable that is fundamentally a function of $Y$. For it to be a valid random variable with respect to the information provided by $Y$, it must be measurable with respect to $\sigma(Y)$.Because the finest non-empty atomic blocks (the smallest building blocks) of $\sigma(Y)$ are the sets $\{\omega_1, \omega_2\}$ and $\{\omega_3, \omega_4\}$, any function that is $\sigma(Y)$-measurable cannot distinguish between elements within the same block. Therefore, $\mathbb{E}[X|Y]$ must take a constant value on $\{\omega_1, \omega_2\}$ and another constant value on $\{\omega_3, \omega_4\}$.

4. Compute $E[X\mid Y]$ explicitly and show that

$$
E[X\mid Y]= \frac{1}{6} \mathbb{1}_{\{Y=0\}} + \frac{1}{2} \mathbb{1}_{\{Y=1\}}
$$


To compute the values over the partitions:For $Y=0$ (on the set $\{\omega_1, \omega_2\}$):$$\mathbb{E}[X|Y=0] = \frac{\mathbb{E}[X \cdot 1_{\{Y=0\}}]}{P(Y=0)} = \frac{X(\omega_1)P(\{\omega_1\}) + X(\omega_2)P(\{\omega_2\})}{P(\{\omega_1\}) + P(\{\omega_2\})}$$$$\mathbb{E}[X|Y=0] = \frac{(0)(0.50) + (1)(0.10)}{0.50 + 0.10} = \frac{0.10}{0.60} = \frac{1}{6}$$For $Y=1$ (on the set $\{\omega_3, \omega_4\}$):$$\mathbb{E}[X|Y=1] = \frac{\mathbb{E}[X \cdot 1_{\{Y=1\}}]}{P(Y=1)} = \frac{X(\omega_3)P(\{\omega_3\}) + X(\omega_4)P(\{\omega_4\})}{P(\{\omega_3\}) + P(\{\omega_4\})}$$$$\mathbb{E}[X|Y=1] = \frac{(0)(0.20) + (1)(0.20)}{0.20 + 0.20} = \frac{0.20}{0.40} = \frac{1}{2}$$Combining these using indicator functions yields:$$\mathbb{E}[X|Y] = \frac{1}{6}1_{\{Y=0\}} + \frac{1}{2}1_{\{Y=1\}}$$

5. Verify the defining property of conditional expectation by checking that for every set $A\in\sigma(Y)$,

$$
\int_A E[X\mid Y]\,dP=\int_A X\,dP.
$$


We must verify that $\int_{A}\mathbb{E}[X|Y]dP = \int_{A}X~dP$ for all four elements in $\sigma(Y)$:Case $A = \emptyset$: Both sides are trivially $0$.Case $A = \Omega$:$\int_{\Omega} X dP = \mathbb{E}[X] = (0)(0.5) + (1)(0.1) + (0)(0.2) + (1)(0.2) = 0.30$$\int_{\Omega}\mathbb{E}[X|Y]dP = \frac{1}{6}P(Y=0) + \frac{1}{2}P(Y=1) = \frac{1}{6}(0.60) + \frac{1}{2}(0.40) = 0.10 + 0.20 = 0.30$Case $A = \{\omega_1, \omega_2\}$:$\int_{A} X dP = X(\omega_1)P(\{\omega_1\}) + X(\omega_2)P(\{\omega_2\}) = 0 + 0.10 = 0.10$$\int_{A}\mathbb{E}[X|Y]dP = \frac{1}{6}P(\{\omega_1, \omega_2\}) = \frac{1}{6}(0.60) = 0.10$Case $A = \{\omega_3, \omega_4\}$:$\int_{A} X dP = X(\omega_3)P(\{\omega_3\}) + X(\omega_4)P(\{\omega_4\}) = 0 + 0.20 = 0.20$$\int_{A}\mathbb{E}[X|Y]dP = \frac{1}{2}P(\{\omega_3, \omega_4\}) = \frac{1}{2}(0.40) = 0.20$The defining property holds exactly across all cases.
6. Find the Conditional PMF $p_{X|Y}(x \mid y)$.

Using the definition $p_{X|Y}(x|y) = \frac{P(X=x, Y=y)}{P(Y=y)}$:For $y=0$:$p_{X|Y}(0|0) = \frac{P(\{\omega_1\})}{P(\{\omega_1, \omega_2\})} = \frac{0.50}{0.60} = \frac{5}{6}$$p_{X|Y}(1|0) = \frac{P(\{\omega_2\})}{P(\{\omega_1, \omega_2\})} = \frac{0.10}{0.60} = \frac{1}{6}$For $y=1$:$p_{X|Y}(0|1) = \frac{P(\{\omega_3\})}{P(\{\omega_3, \omega_4\})} = \frac{0.20}{0.40} = \frac{1}{2}$$p_{X|Y}(1|1) = \frac{P(\{\omega_4\})}{P(\{\omega_3, \omega_4\})} = \frac{0.20}{0.40} = \frac{1}{2}$

7. Verify the identity:
$$\mathbb{E}[X\mid Y=y] = \sum_{x \in \{0,1\}} x \cdot p_{X|Y}(x \mid y)$$.

If $y=0$: $\sum_{x\in\{0,1\}} x \cdot p_{X|Y}(x|0) = 0\cdot(\frac{5}{6}) + 1\cdot(\frac{1}{6}) = \frac{1}{6}$ (Matches Step 4)If $y=1$: $\sum_{x\in\{0,1\}} x \cdot p_{X|Y}(x|1) = 0\cdot(\frac{1}{2}) + 1\cdot(\frac{1}{2}) = \frac{1}{2}$ (Matches Step 4)

8. Calculate the **Mean Square Error (MSE)**

The MSE when using $\mathbb{E}[X|Y]$ as our estimator is the expected conditional variance:$$\text{MSE} = \mathbb{E}[(X - \mathbb{E}[X|Y])^2] = \sum_{\omega \in \Omega} (X(\omega) - \mathbb{E}[X|Y](\omega))^2 P(\{\omega\})$$$$\text{MSE} = \left(0-\frac{1}{6}\right)^2(0.5) + \left(1-\frac{1}{6}\right)^2(0.1) + \left(0-\frac{1}{2}\right)^2(0.2) + \left(1-\frac{1}{2}\right)^2(0.2)$$$$\text{MSE} = \left(\frac{1}{36}\right)(0.5) + \left(\frac{25}{36}\right)(0.1) + \left(\frac{1}{4}\right)(0.2) + \left(\frac{1}{4}\right)(0.2)$$$$\text{MSE} = \frac{0.5 + 2.5}{36} + \frac{0.2 + 0.2}{4} = \frac{3}{36} + \frac{0.4}{4} = \frac{1}{12} + \frac{1}{10} = \frac{11}{60} \approx 0.1833$$

9. How would the **Mean Square Error (MSE)** change if the bank decided to ignore the customer risk class $Y$ and instead used the global average fraud rate ($E[X] = 0.30$) as a constant estimate for every transaction? Calculate this error and explain its relationship to the **Law of Total Variance**.

If the bank ignores $Y$ and uses the global average constant estimate $\mathbb{E}[X] = 0.30$, the new error is the total variance of $X$:$$\text{MSE}_{\text{constant}} = \text{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2$$Since $X$ is binary, $X^2 = X$, so $\mathbb{E}[X^2] = \mathbb{E}[X] = 0.30$.$$\text{Var}(X) = 0.30 - (0.30)^2 = 0.30 - 0.09 = 0.21$$Relationship to the Law of Total Variance:The Law of Total Variance states:$$\text{Var}(X) = \mathbb{E}[\text{Var}(X|Y)] + \text{Var}(\mathbb{E}[X|Y])$$$\mathbb{E}[\text{Var}(X|Y)]$ is the MSE with information $Y$ ($0.1833$).$\text{Var}(\mathbb{E}[X|Y])$ represents the variance explained by the information $Y$.Let's compute $\text{Var}(\mathbb{E}[X|Y])$ to close the loop:$$\mathbb{E}[(\mathbb{E}[X|Y])^2] = \left(\frac{1}{6}\right)^2(0.6) + \left(\frac{1}{2}\right)^2(0.4) = \frac{1}{36}(0.6) + \frac{1}{4}(0.4) = \frac{1}{60} + \frac{1}{10} = \frac{7}{60}$$$$\text{Var}(\mathbb{E}[X|Y]) = \mathbb{E}[(\mathbb{E}[X|Y])^2] - (\mathbb{E}[X])^2 = \frac{7}{60} - 0.09 = 0.1167 - 0.09 = 0.0267$$Verification: $\text{MSE}_{\text{with } Y} + \text{Var}(\mathbb{E}[X|Y]) = \frac{11}{60} + \frac{1.6}{60} = \frac{12.6}{60} = 0.21 = \text{Var}(X)$.Ignoring $Y$ increases our error from $0.1833$ to $0.21$ because we lose the predictive power/variance explained by the risk tier.


10. Using the joint distribution of $(X, Y)$ derived from the sample space:

    a) Compute the entropy $H(X)$, representing the bank's initial uncertainty regarding fraud.

    $$H(X) = - [0.7 \log_2(0.7) + 0.3 \log_2(0.3)] \approx -[-0.3653 - 0.5211] = 0.8813\text{ bits}$$

    b) Compute the conditional entropy $H(X \mid Y)$, representing the uncertainty remaining after the customer's risk class is revealed.

    $$H(X|Y) = P(Y=0)H(X|Y=0) + P(Y=1)H(X|Y=1)$$$$H(X|Y=0) = -\left[\frac{5}{6}\log_2\left(\frac{5}{6}\right) + \frac{1}{6}\log_2\left(\frac{1}{6}\right)\right] \approx 0.6500\text{ bits}$$$$H(X|Y=1) = -\left[\frac{1}{2}\log_2\left(\frac{1}{2}\right) + \frac{1}{2}\log_2\left(\frac{1}{2}\right)\right] = 1.0000\text{ bits}$$$$H(X|Y) = (0.6)(0.6500) + (0.4)(1.0000) = 0.3900 + 0.4000 = 0.7900\text{ bits}$$

    c) Calculate the Mutual Information $I(X; Y)$.

    $$I(X;Y) = H(X) - H(X|Y) = 0.8813 - 0.7900 = 0.0913\text{ bits}$$

    d) Based on these values, what percentage of the total uncertainty about fraud is removed by knowing the customer's risk class?
$$\%\text{ Uncertainty Removed} = \frac{I(X;Y)}{H(X)} \times 100 = \frac{0.0913}{0.8813} \times 100 \approx 10.36\%$$

# Q2) Conditional Density of the Multivariate Gaussian

Let $\mathbf{Z}$ be the concatenated random matrix
$\mathbf{Z} = \begin{bmatrix} \mathbf{X} \\ \mathbf{Y} \end{bmatrix} \in \mathbb{R}^{n+m}$.


The random variable $\mathbf{Z}$ is said to have a **joint multivariate normal distribution** if every linear combination of its components follows a univariate normal distribution.

For $\mathbf{Z}$ to possess a **joint density function** (meaning its distribution is absolutely continuous with respect to the Lebesgue measure on $\mathbb{R}^{n+m}$), its covariance matrix must be non-singular (positive definite).


The joint density function $f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z})$ for $\mathbf{z} \in \mathbb{R}^{n+m}$ is given by:

$$f_{\mathbf{X}, \mathbf{Y}}(\mathbf{z}) = \frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu}) \right).$$

Here:

* **Mean ($\boldsymbol{\mu}$):** $\boldsymbol{\mu} = E[\mathbf{Z}] = \begin{bmatrix} \boldsymbol{\mu}_X \\ \boldsymbol{\mu}_Y \end{bmatrix} \in \mathbb{R}^{n+m}$, where $\boldsymbol{\mu}_X \in \mathbb{R}^n$ and $\boldsymbol{\mu}_Y \in \mathbb{R}^m$.
* **Covariance Matrix ($\boldsymbol{\Sigma}$):** A symmetric, positive definite matrix $\boldsymbol{\Sigma} \in \mathbb{R}^{(n+m) \times (n+m)}$ partitioned as:
$$\boldsymbol{\Sigma} = \begin{bmatrix} \boldsymbol{\Sigma}_{XX} & \boldsymbol{\Sigma}_{XY} \\ \boldsymbol{\Sigma}_{YX} & \boldsymbol{\Sigma}_{YY} \end{bmatrix}$$
    * $\boldsymbol{\Sigma}_{XX} = \text{Var}(\mathbf{X})$ (size $n \times n$)
    * $\boldsymbol{\Sigma}_{YY} = \text{Var}(\mathbf{Y})$ (size $m \times m$)
    * $\boldsymbol{\Sigma}_{XY} = \text{Cov}(\mathbf{X}, \mathbf{Y}) = \boldsymbol{\Sigma}_{YX}^T$ (size $n \times m$)
* **Determinant ($|\boldsymbol{\Sigma}|$):** The normalization factor ensuring the total integral is 1.

1. Show that the conditional distribution is given by $f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$

$$f_{\mathbf{X}|\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\mathbf{S}_{XX}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right),$$
where
* $\boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_x + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_y)$


To show that $f_{X|Y}(x|y) = \frac{f_{X,Y}(x,y)}{f_Y(y)}$ takes the requested form, we use a multivariate transformation strategy based on completing the square or constructing an independent linear combination.Let's define a new variable $W = X - \Sigma_{XY}\Sigma_{YY}^{-1}Y$.Because $W$ and $Y$ are linear transformations of the jointly Gaussian vector $Z$, they are themselves jointly Gaussian. Let's look at their cross-covariance:$$\text{Cov}(W, Y) = \text{Cov}(X - \Sigma_{XY}\Sigma_{YY}^{-1}Y, Y) = \text{Cov}(X,Y) - \Sigma_{XY}\Sigma_{YY}^{-1}\text{Var}(Y)$$$$\text{Cov}(W, Y) = \Sigma_{XY} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YY} = \Sigma_{XY} - \Sigma_{XY} = 0$$Since $W$ and $Y$ are jointly Gaussian and uncorrelated, they are statistically independent.Mean of $W$: $\mu_W = \mathbb{E}[X] - \Sigma_{XY}\Sigma_{YY}^{-1}\mathbb{E}[Y] = \mu_X - \Sigma_{XY}\Sigma_{YY}^{-1}\mu_Y$Variance of $W$ ($S_{XX}$ or Schur Complement):$$\Sigma_W = \text{Var}(X - \Sigma_{XY}\Sigma_{YY}^{-1}Y) = \text{Var}(X) - \Sigma_{XY}\Sigma_{YY}^{-1}\text{Cov}(Y,X) - \text{Cov}(X,Y)\Sigma_{YY}^{-1}\Sigma_{YX} + \Sigma_{XY}\Sigma_{YY}^{-1}\text{Var}(Y)\Sigma_{YY}^{-1}\Sigma_{YX}$$$$\Sigma_W = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX} + \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX} = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX} = \Sigma_{X|Y}$$Because $W$ is independent of $Y$, the conditional distribution of $W$ given $Y=y$ is exactly equal to its marginal distribution:$$W|Y=y \sim \mathcal{N}(\mu_W, \Sigma_{X|Y})$$Now back-substitute to find $X|Y=y$. Since $X = W + \Sigma_{XY}\Sigma_{YY}^{-1}Y$, when $Y$ is fixed to a deterministic value $y$:$$X|Y=y = W + \Sigma_{XY}\Sigma_{YY}^{-1}y$$Thus, the conditional mean $\mu^*$ is:$$\mu^* = \mathbb{E}[W] + \Sigma_{XY}\Sigma_{YY}^{-1}y = \mu_X - \Sigma_{XY}\Sigma_{YY}^{-1}\mu_Y + \Sigma_{XY}\Sigma_{YY}^{-1}y = \mu_X + \Sigma_{XY}\Sigma_{YY}^{-1}(y - \mu_Y)$$The variance structure remains unaltered by adding a deterministic shift:$$\text{Var}(X|Y=y) = \text{Var}(W) = \Sigma_{X|Y}$$Plugging this mean vector $\mu^*$ and covariance matrix $\Sigma_{X|Y}$ into the general template of an $n$-dimensional multivariate normal density function yields:$$f_{X|Y}(x|y) = \frac{1}{\sqrt{(2\pi)^{n}|\Sigma_{X|Y}|}}\exp\left(-\frac{1}{2}(x-\mu^{*})^{T}\Sigma_{X|Y}^{-1}(x-\mu^{*})\right)$$(Where $|\Sigma_{X|Y}|$ maps directly to the designated notation $|S_{XX}|$ for the Schur complement of the block matrix).

# Q3) Continuous Fault Monitoring of a Spindle Motor

A factory monitors the health of a high-speed CNC spindle. The critical variable is the **Internal Temperature ($X$)**, but because the motor is sealed, they can only observe the **Total Spindle Power ($Y$)**.

**The Setup:**
1.  **Hidden Temperature ($X$):** The standardized temperature deviation follows a standard normal distribution:
    $$X \sim \mathscr{N}(0, 1)$$
2.  **Mechanical Load ($Q$):** The torque required to cut material creates "noise" in the power reading. It is independent of temperature and follows:
    $$Q \sim \mathscr{N}(0, 4)$$
3.  **The Measurement ($Y$):** The power sensor $Y$ measures the combined effect of thermal load and mechanical torque:
$$Y = X + Q$$

---

**Answer the followings:**

**1. Joint Distribution**
Identify the distribution of the random varaible $(X, Y)$.
* Show that $Y \sim \mathscr{N}(0, 5)$.
* Find the covariance $\text{Cov}(X, Y)$.

We are given $X \sim \mathcal{N}(0,1)$ and $Q \sim \mathcal{N}(0,4)$, where $X$ and $Q$ are independent. $Y = X + Q$.Mean of $Y$: $\mathbb{E}[Y] = \mathbb{E}[X] + \mathbb{E}[Q] = 0 + 0 = 0$Variance of $Y$: Since $X$ and $Q$ are independent:$$\text{Var}(Y) = \text{Var}(X) + \text{Var}(Q) = 1 + 4 = 5 \implies Y \sim \mathcal{N}(0,5)$$Covariance $\text{Cov}(X,Y)$:$$\text{Cov}(X, Y) = \text{Cov}(X, X + Q) = \text{Var}(X) + \text{Cov}(X, Q) = 1 + 0 = 1$$Because linear combinations of $X$ and $Y$ are univariate normals, they follow a joint bivariate normal distribution:$$\begin{bmatrix} X \\ Y \end{bmatrix} \sim \mathcal{N}\left( \begin{bmatrix} 0 \\ 0 \end{bmatrix}, \begin{bmatrix} 1 & 1 \\ 1 & 5 \end{bmatrix} \right)$$

**2. Information Geometry**
Explain why $X$ is **not** $\sigma(Y)$-measurable. In practical terms, if the power meter shows $Y = 5$, why can't the engineer know for certain if the motor is overheating ($X$ is high) or if the material being cut is simply very hard ($Q$ is high)?

$X$ is not $\sigma(Y)$-measurable. For $X$ to be $\sigma(Y)$-measurable, knowing the value of $Y$ must uniquely pinpoint the value of $X$ (there must exist a deterministic function $g$ such that $X = g(Y)$).In practical terms, if $Y = 5$, the engine can be governed by an infinite distribution of combinations because $Y$ is a single combined measurement of two distinct underlying latent variables ($X$ and $Q$). A high power reading of $5$ could be caused by an overheated motor running normally ($X=4, Q=1$) or a completely cold motor cutting an exceptionally dense material ($X=-1, Q=6$). The measurement system experiences information loss due to compounding noise, making absolute inversion impossible.

**3. Conditional Expectation (The Best Estimate)**
Compute the conditional expectation $E[X \mid Y]$.
* Using the property for joint normals: $E[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$.
* Express this as a "signal-to-total-variance" ratio.

Using the provided property for joint normals:$$\mathbb{E}[X|Y=y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y) = 0 + \frac{1}{5}(y - 0) = \frac{1}{5}y$$Signal-to-Total-Variance Ratio Expression:$$\mathbb{E}[X|Y] = \left(\frac{\sigma_X^2}{\sigma_X^2 + \sigma_Q^2}\right) Y = \left(\frac{1}{1 + 4}\right) Y = \frac{1}{5}Y$$

**4. Verification of the Residual**
Let $\widehat{X} = E[X \mid Y]$. Define the estimation error (residual) as $Z = X - \widehat{X}$.
* Prove that $E[Z] = 0$.
* Verify that the error $Z$ is uncorrelated with the observation $Y$ (the Orthogonality Principle).

Let $\hat{X} = \mathbb{E}[X|Y] = \frac{1}{5}Y$. The residual is $Z = X - \hat{X} = X - \frac{1}{5}Y$.Prove $\mathbb{E}[Z] = 0$:$$\mathbb{E}[Z] = \mathbb{E}[X] - \frac{1}{5}\mathbb{E}[Y] = 0 - 0 = 0$$Verify Orthogonality ($\text{Cov}(Z,Y) = 0$):$$\text{Cov}(Z, Y) = \text{Cov}\left(X - \frac{1}{5}Y, Y\right) = \text{Cov}(X,Y) - \frac{1}{5}\text{Var}(Y) = 1 - \frac{1}{5}(5) = 1 - 1 = 0$$Since their covariance is $0$, the residual error is completely uncorrelated with our observation space.

**5. Information Theory: Differential Entropy**
* Compute the differential entropy $h(X)$.
* Compute the conditional differential entropy $h(X \mid Y)$.
* *Note: The variance of the conditional distribution $(X \mid Y=y)$ is given by $\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)}$.*

The differential entropy of a continuous Gaussian distribution $\mathcal{N}(\mu, \sigma^2)$ is given by $h = \frac{1}{2}\ln(2\pi e \sigma^2)$.Differential Entropy $h(X)$:$$h(X) = \frac{1}{2}\ln(2\pi e (1)) = \frac{1}{2}\ln(2\pi e)$$Conditional Differential Entropy $h(X|Y)$:First calculate the conditional variance $\sigma_{X|Y}^2$:$$\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)} = 1 - \frac{1^2}{5} = \frac{4}{5} = 0.8$$$$h(X|Y) = \frac{1}{2}\ln\left(2\pi e \cdot \frac{4}{5}\right)$$

**6. Mutual Information**
Calculate $I(X; Y) = h(X) - h(X \mid Y)$.
* What percentage of the "uncertainty" about the temperature is removed by observing the power meter?
* How would this change if the mechanical noise $Q$ had a variance of $100$ instead of $4$?

$$I(X;Y) = h(X) - h(X|Y) = \frac{1}{2}\ln(2\pi e) - \frac{1}{2}\ln\left(2\pi e \cdot \frac{4}{5}\right) = \frac{1}{2}\ln\left(\frac{2\pi e}{2\pi e \cdot \frac{4}{5}}\right) = \frac{1}{2}\ln\left(\frac{5}{4}\right) \approx 0.1116\text{ nats}$$Percentage of uncertainty removed:$$\frac{I(X;Y)}{h(X)} \text{ cannot be framed directly as a percentage like discrete entropy since } h(X) \text{ scales relative to coordinates.}$$However, looking at variance drop: our baseline variance was $1.0$, and it dropped to $0.8$. Thus, $20\%$ of the structural variance is eliminated by observing the power meter.If mechanical noise variance $\sigma_Q^2 = 100$:$$\text{New } \text{Var}(Y) = 1 + 100 = 101, \quad \text{Cov}(X,Y) = 1$$$$\sigma_{X|Y}^2 = 1 - \frac{1}{101} = \frac{100}{101} \approx 0.9901$$$$I(X;Y) = \frac{1}{2}\ln\left(\frac{1}{100/101}\right) = \frac{1}{2}\ln\left(\frac{101}{100}\right) \approx 0.00497\text{ nats}$$The variance eliminated drops from $20\%$ down to roughly $0.99\%$, showing that overwhelming noise nearly zeroes out the predictive value of the instrument.



**7. Diagnostic Interpretation**
The factory manager says: *"Since $Y = X + Q$, our best estimate of temperature should just be $Y$ itself."*
Using your result from Question 3, explain why $Y$ is a **biased** estimator of $X$ and why $E[X \mid Y]$ "shrinks" the measurement toward the mean (zero).


  The factory manager's statement is incorrect. If you use $Y$ directly as an estimator for $X$, its expected value given $X$ is $\mathbb{E}[Y|X] = \mathbb{E}[X+Q|X] = X$. While it is unbiased in that directional sense, its absolute variance scales unsustainably ($\text{Var}(Y) = 5$, whereas $\text{Var}(X) = 1$). If you observe $Y=5$, guessing $X=5$ overshoots wildly because a major portion of that value likely belongs to the noise component $Q$.The conditional expectation $\mathbb{E}[X|Y] = \frac{1}{5}Y$ accounts for this by performing a shrinkage effect toward the mean (zero). It weights the measurement by the signal-to-total-variance ratio. When a high measurement is observed, it recognizes that only a fraction of the variation is statistically likely to stem from the core signal ($X$), dynamically pulling the best guess down to safe structural bounds.
