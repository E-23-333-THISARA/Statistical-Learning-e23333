#Q3) Transformer Failiures

In a certain electricity distribution district, the number of emergency transformer failures reported in a single day is modeled as a discrete random variable $X$. Assume that $X$ takes values in the set $\{0,1,2,3\}$, with probabilities

$$
P(X=0)=0.50,\qquad P(X=1)=0.30,\qquad P(X=2)=0.15,\qquad P(X=3)=0.05.
$$

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.
2. Define $X$ as a random variable on this space.
3. Determine the probability mass function of $X$.
4. Compute the expectation $E[X]$.
5. Compute the variance $\mathrm{Var}(X)$.
6. Find $P(X\ge 1)$ and $P(X\ge 2)$.
7. Briefly explain the practical meaning of the expectation and variance in this setting.



    
problem models discrete emergency transformer failures $X \in \{0, 1, 2, 3\}$.     
1. Probability Space $(\Omega, \mathcal{F}, P)$:  Sample Space: $\Omega = \{0, 1, 2, 3\}$, representing the possible number of failures.  $\sigma$-algebra: $\mathcal{F} = \mathcal{P}(\Omega)$ (the power set), containing all $2^4 = 16$ possible events.Probability Measure: $P$ is defined by the assigned weights: $P(\{0\})=0.50$, $P(\{1\})=0.30$, $P(\{2\})=0.15$. Since probabilities must sum to 1, $P(\{3\}) = 1 - (0.50 + 0.30 + 0.15) = 0.05$.
 2. Define X as a Random Variable:
$X$ is a functional mapping $X: \Omega \to \mathbb{R}$ such that $X(\omega) = \omega$ for all $\omega \in \Omega$.  
3. Probability Mass Function (PMF):
  $$p_X(x) = \begin{cases} 0.50, & x=0 \\ 0.30, & x=1 \\ 0.15, & x=2 \\ 0.05, & x=3 \\ 0, & \text{otherwise} \end{cases}$$
  4. Expectation $E[X]$:
$E[X] = \sum x \cdot P(X=x) = (0 \times 0.50) + (1 \times 0.30) + (2 \times 0.15) + (3 \times 0.05) = 0 + 0.30 + 0.30 + 0.15 = \mathbf{0.75}$ failures/day.  

5. Variance $Var(X)$:
First, find $E[X^2] = (0^2 \times 0.50) + (1^2 \times 0.30) + (2^2 \times 0.15) + (3^2 \times 0.05) = 0 + 0.30 + 0.60 + 0.45 = 1.35$.
$Var(X) = E[X^2] - (E[X])^2 = 1.35 - (0.75)^2 = 1.35 - 0.5625 = \mathbf{0.7875}$.
 6. Probabilities $P(X \ge 1)$ and $P(X \ge 2)$:  $P(X \ge 1) = 1 - P(X=0) = 1 - 0.50 = \mathbf{0.50}$.$P(X \ge 2) = P(X=2) + P(X=3) = 0.15 + 0.05 = \mathbf{0.20}$.
 7. Practical Meaning:  Expectation (0.75): The long-run average number of failures per day the district should anticipate.Variance (0.7875): A measure of the "spread" or uncertainty; it indicates how much the daily failure rate fluctuates around the average.

# Q4) Probability of Defective Items

In a manufacturing process, the proportion of defective items in a large batch produced on a given day is modeled by a continuous random variable $X$. Since $X$ represents a proportion, it takes values in the interval $[0,1]$. Suppose that $X$ has a Beta distribution with parameters $\alpha=2$ and $\beta=18$, so that its density is

$$
f(x)=
\begin{cases}
\dfrac{1}{B(2,18)}x^{1}(1-x)^{17}, & 0<x<1,\\\
0, & \text{otherwise},
\end{cases}
$$

where $B(2,18)$ is the Beta function.

Answer the following:

1. Construct a probability space $(\Omega,\mathfrak F,P)$ suitable for this model.

2. Define $X$ as a continuous random variable on this space.

3. Write down the probability density function of $X$.

4. Verify that $f$ is a valid probability density function.

5. Compute the expectation $E[X]$.

6. Compute the variance $\mathrm{Var}(X)$.

7. Find the probability that the defective proportion is at most $0.10$, that is, compute

$$
P(X\le 0.10).
$$

You may leave your answer in Beta integral form.

8. Find the probability that the defective proportion lies between $0.05$ and $0.15$, that is, compute

$$
P(0.05\le X\le 0.15).
$$

You may leave your answer in integral form.

9. Briefly explain why a Beta distribution is a reasonable model for this situation.

10. Interpret the expectation and variance in the context of production quality.


Modeling a proportion $X \in [0, 1]$ using a Beta Distribution ($\alpha=2, \beta=18$).
 1. Probability Space $(\Omega, \mathcal{F}, P)$:  $\Omega = [0, 1]$.$\mathcal{F} = \mathcal{B}([0, 1])$, the Borel $\sigma$-algebra on the unit interval.$P$ is defined by the density $f(x)$.  
2. Random Variable Definition:
$X: \Omega \to [0, 1]$ is the identity mapping $X(\omega) = \omega$, representing the defective proportion.
 3. Probability Density Function (PDF):
$f(x) = \frac{x^{2-1}(1-x)^{18-1}}{B(2,18)} = \frac{x(1-x)^{17}}{B(2,18)}$ for $0 < x < 1$.
 4. Verification:
A valid PDF must integrate to 1. By definition of the Beta function:
$\int_{0}^{1} x^{\alpha-1}(1-x)^{\beta-1} dx = B(\alpha, \beta)$.
Thus, $\int_{0}^{1} \frac{x(1-x)^{17}}{B(2,18)} dx = \frac{1}{B(2,18)} \int_{0}^{1} x^{2-1}(1-x)^{18-1} dx = \frac{B(2,18)}{B(2,18)} = 1$.
 5. Expectation $E[X]$:
For Beta($\alpha, \beta$): $E[X] = \frac{\alpha}{\alpha+\beta} = \frac{2}{2+18} = \mathbf{0.10}$.
 6. Variance $Var(X)$:
$Var(X) = \frac{\alpha\beta}{(\alpha+\beta)^2(\alpha+\beta+1)} = \frac{2 \times 18}{(20)^2(21)} = \frac{36}{400 \times 21} = \frac{36}{8400} \approx \mathbf{0.00428}$.
 7. $P(X \le 0.10)$:
$P(X \le 0.10) = \frac{1}{B(2,18)} \int_{0}^{0.10} x(1-x)^{17} dx$.
 8. $P(0.05 \le X \le 0.15)$:
$P(0.05 \le X \le 0.15) = \frac{1}{B(2,18)} \int_{0.05}^{0.15} x(1-x)^{17} dx$.
 9. Why Beta Distribution?:
The Beta distribution is ideal for modeling proportions because its support is strictly bounded between 0 and 1, and its shape parameters ($\alpha, \beta$) allow for flexible modeling of skewness (in this case, a low defective rate).  
10. Interpretation:  Expectation (0.10): On average, 10% of items in a batch are expected to be defective.Variance: Low variance indicates high consistency in the manufacturing process quality.

# Q5) Expected Project Completion Time

A construction company is estimating the time required to install a backup generator at a hospital. Let $X$ denote the number of hours needed to complete the installation. Based on expert judgment, the shortest possible time is $4$ hours, the longest possible time is $10$ hours, and the most likely completion time is $6$ hours. Assume that $X$ follows a Triangular distribution with parameters $a=4$, $b=10$, and $c=6$.

Answer the following:

1. Specify the support of $X$.

2. Write down the probability density function $f_X(x)$.

3. Verify that $f_X$ integrates to $1$ over its support.

4. Compute the expectation $E[X]$.

5. Compute the variance $\mathrm{Var}(X)$.

6. Interpret the values of $a$, $b$, and $c$ in the context of this problem.

7. Explain briefly why a Triangular distribution is a reasonable model in this setting.

8. Compute the probability that the installation takes at most $5$ hours, that is, compute

$$
P(X \leq 5).
$$

9. Compute the probability that the installation takes between $5$ and $8$ hours, that is, compute

$$
P(5 \leq X \leq 8).
$$

10. Briefly explain how this model could be used in project planning or sensitivity analysis.


problem involves a Triangular distribution, commonly used in project management (like the Three-Point Estimation technique) when data is scarce but expert judgment provides the limits and the most likely outcome.
1. Specify the support of XThe support of a distribution is the set of values for which the probability density is non-zero. For this backup generator installation, the support is the interval between the shortest and longest possible times:Support: $[4, 10]$.
  2. Write down the probability density function $f_X(x)$
The PDF for a Triangular distribution with $a=4, b=10,$ and $c=6$ is defined piecewise:  $$f_X(x) = \begin{cases}\frac{2(x-a)}{(b-a)(c-a)} = \frac{2(x-4)}{(10-4)(6-4)} = \mathbf{\frac{x-4}{6}}, & 4 \le x < 6 \\frac{2(b-x)}{(b-a)(b-c)} = \frac{2(10-x)}{(10-4)(10-6)} = \mathbf{\frac{10-x}{12}}, & 6 \le x \le 10 \0, & \text{otherwise}\end{cases}$$
3. Verify that $f_X$ integrates to 1 over its support
To verify this, we calculate the area under the two linear segments:
  $$\int_{4}^{10} f_X(x)dx = \int_{4}^{6} \frac{x-4}{6} dx + \int_{6}^{10} \frac{10-x}{12} dx$$$$= \left[ \frac{(x-4)^2}{12} \right]_4^6 + \left[ -\frac{(10-x)^2}{24} \right]_6^{10} = \left(\frac{4}{12} - 0\right) + \left(0 - \left(-\frac{16}{24}\right)\right) = \frac{1}{3} + \frac{2}{3} = \mathbf{1}$$
  4. Compute the expectation $E[X]$
The mean of a Triangular distribution is the simple average of its three parameters:  $E[X] = \frac{a+b+c}{3} = \frac{4+10+6}{3} = \frac{20}{3} \approx \mathbf{6.67}$ hours.
5. Compute the variance $Var(X)$
Using the variance formula for this distribution:  $Var(X) = \frac{a^2+b^2+c^2-ab-bc-ca}{18} = \frac{4^2+10^2+6^2 - (40+60+24)}{18} = \frac{152-124}{18} = \frac{28}{18} \approx \mathbf{1.56}$.
6. Interpret the values of a, b, and c$a=4$: The optimistic time; the minimum hours required if everything goes perfectly.  $b=10$: The pessimistic time; the maximum hours required if significant delays occur.  $c=6$: The most likely time; the mode or the peak of the probability density.  
7. Why is the Triangular distribution a reasonable model?
It is useful when actual historical data is missing but experts can provide "best-case," "worst-case," and "most-likely" estimates. It is simpler to calculate than the Beta distribution while still accounting for skewness (since $c$ is not exactly in the middle of $a$ and $b$).
 8. Compute the probability $P(X \le 5)$
Since $x=5$ falls in the first interval ($4 \le x < 6$), we integrate the first part of the PDF:  $P(X \le 5) = \int_{4}^{5} \frac{x-4}{6} dx = \left[ \frac{(x-4)^2}{12} \right]_4^5 = \frac{(5-4)^2}{12} - 0 = \mathbf{\frac{1}{12}} \approx \mathbf{0.0833}$ (or $8.33\%$).
9. Compute the probability $P(5 \le X \le 8)$
This spans across the mode ($c=6$), so we sum the area from 5 to 6 and 6 to 8:  $P(5 \le X \le 8) = \int_{5}^{6} \frac{x-4}{6} dx + \int_{6}^{8} \frac{10-x}{12} dx$$= \left[ \frac{(x-4)^2}{12} \right]_5^6 + \left[ -\frac{(10-x)^2}{24} \right]_6^8$$= (\frac{4}{12} - \frac{1}{12}) + (-\frac{4}{24} - (-\frac{16}{24})) = \frac{3}{12} + \frac{12}{24} = 0.25 + 0.5 = \mathbf{0.75}$ (or $75\%$).
10. Usage in project planning or sensitivity analysis
This model allows project managers to quantify risks. By calculating the probability of finishing within certain timeframes (like the $75\%$ chance calculated above), they can set more realistic deadlines and determine which tasks have the most "slack" or "sensitivity" to delays.

#Q6) Expected Project Completion Time Using PERT

A project manager is estimating the time required to complete a critical software integration task. Let $X$ denote the number of days required to complete the task. Based on expert judgment, the shortest possible time is $a=8$ days, the longest possible time is $b=20$ days, and the most likely completion time is $c=11$ days. Assume that $X$ follows a PERT distribution with parameters $(a,b,c)$.

The PERT distribution is defined as a Beta distribution scaled to the interval $[a,b]$, with internal shape parameters

$$
\alpha = 1 + 4\left( \frac{c-a}{b-a} \right)
$$

and

$$
\beta = 1 + 4\left( \frac{b-c}{b-a} \right).
$$

Its probability density function is

$$
f(x)=\frac{(x-a)^{\alpha-1}(b-x)^{\beta-1}}{B(\alpha,\beta)(b-a)^{\alpha+\beta-1}}, \qquad a \leq x \leq b,
$$

where

$$
B(\alpha,\beta)=\int_0^1 t^{\alpha-1}(1-t)^{\beta-1},dt.
$$

Answer the following:

1. Compute the values of the shape parameters $\alpha$ and $\beta$.

2. Specify the support of $X$.

3. Write down the probability density function $f_X(x)$ for this model.

4. State the probability space $(\Omega,\mathfrak F,P)$ suitable for this continuous model.

5. Define $X$ as a random variable on this space.

6. Show that the expectation $E[X]$ is given by

$$
E[X]=\frac{a+4c+b}{6}.
$$

7. Compute the variance $\mathrm{Var}(X)$ using the approximation

$$
\mathrm{Var}(X)=\frac{(E[X]-a)(b-E[X])}{7}.
$$

8. Compute the approximate standard deviation $\sigma$.

9. Briefly explain why a PERT distribution may be more appropriate than a Triangular distribution in this setting byt comparing the two results.

10. Interpret the practical meaning of the values $a$, $b$, $c$, $E[X]$, and $\sigma$ in the context of project planning.


The PERT (Program Evaluation and Review Technique) distribution is a variation of the Beta distribution often preferred over the Triangular distribution because it produces a smoother curve that assigns less probability to the extreme tails (the "optimistic" and "pessimistic" ends).For this problem, the parameters are:Minimum (a): 8 days   Maximum (b): 20 days   Most Likely (c): 11 days  
 1. Compute the values of the shape parameters $\alpha$ and $\beta$ The PERT distribution uses these internal shape parameters to define the curve based on the "4-weight" mode logic:  $\alpha = 1 + 4\left(\frac{c-a}{b-a}\right) = 1 + 4\left(\frac{11-8}{20-8}\right) = 1 + 4\left(\frac{3}{12}\right) = 1 + 1 = \mathbf{2}$   $\beta = 1 + 4\left(\frac{b-c}{b-a}\right) = 1 + 4\left(\frac{20-11}{20-8}\right) = 1 + 4\left(\frac{9}{12}\right) = 1 + 3 = \mathbf{4}$   
 2. Specify the support of X   The support is the interval of possible completion times: $[8, 20]$.
  3. Write down the probability density function $f_X(x)$ Using the provided formula  and substituting $\alpha=2, \beta=4, a=8, b=12$:  $$f(x) = \frac{(x-8)^{2-1}(20-x)^{4-1}}{B(2,4)(20-8)^{2+4-1}} = \mathbf{\frac{(x-8)(20-x)^3}{B(2,4)(12)^5}}, \text{ for } 8 \le x \le 20$$
  4. State the probability space $(\Omega, \mathcal{F}, P)$   Sample Space ($\Omega$): $[8, 20]$, representing the range of possible days.  $\sigma$-algebra ($\mathcal{F}$): $\mathcal{B}([8, 20])$, the Borel $\sigma$-algebra on the interval.  Probability Measure ($P$): Defined by the integral of the PDF $f_X(x)$ over any measurable subset of $\Omega$.
    
  5. Define X as a random variable   $X$ is the identity mapping $X: \Omega \to [8, 20]$, where $X(\omega) = \omega$ represents the number of days required to complete the software integration task.
6. Show that the expectation $E[X]$ is given by $\frac{a+4c+b}{6}$
The mean of a Beta distribution on $[0,1]$ is $\mu = \frac{\alpha}{\alpha+\beta}$. For a PERT distribution scaled to $[a,b]$, the mean is $a + (b-a)\frac{\alpha}{\alpha+\beta}$.
Substituting our $\alpha$ and $\beta$ formulas:
  $$E[X] = a + (b-a)\frac{1+4(\frac{c-a}{b-a})}{[1+4(\frac{c-a}{b-a})] + [1+4(\frac{b-c}{b-a})]} = a + (b-a)\frac{1+4(\frac{c-a}{b-a})}{6} = \frac{6a + (b-a) + 4(c-a)}{6}$$$$E[X] = \frac{6a + b - a + 4c - 4a}{6} = \mathbf{\frac{a + 4c + b}{6}}$$Calculation for this task: $\frac{8 + 4(11) + 20}{6} = \frac{72}{6} = \mathbf{12}$ days.
  7. Compute the variance $Var(X)$ using the approximation   $Var(X) = \frac{(E[X]-a)(b-E[X])}{7} = \frac{(12-8)(20-12)}{7} = \frac{4 \times 8}{7} = \frac{32}{7} \approx \mathbf{4.57}$.
   8. Compute the approximate standard deviation $\sigma$   $\sigma = \sqrt{Var(X)} = \sqrt{4.57} \approx \mathbf{2.14}$ days.
   9. PERT vs. Triangular Distribution   Triangular Mean: $\frac{8+20+11}{3} = 13$ days.PERT Mean: $12$ days.Comparison: The PERT distribution is often more appropriate because it places more emphasis on the "most likely" value ($c$) by giving it 4 times the weight of the extremes. This results in a distribution that is less sensitive to extreme "outlier" estimates for $a$ and $b$ than the Triangular model.
  10. Interpret practical meanings   $a, b, c$: Represent the range of expert uncertainty (best case, worst case, mode).  $E[X]$: The "weighted average" duration used for critical path scheduling.  $\sigma$: Represents the risk or uncertainty; a larger $\sigma$ indicates a more unpredictable task
